GENETIC ALGORITHMS
-> create random solutions
-> select the best solutions
-> cross over
-> mutation
-> repeat the process till you wont get best solutions

```python
import random
def marks(x,y,z):
    return 3*x**3+4*y**4+z**2-200

def fitness(x,y,z):
    ans=marks(x,y,z)
    if ans==0:
        return 99999
    else:
        return abs(1/ans)

sol=[]        
#random solutions
for i in range(1000):
    a=random.uniform(0,100)
    b=random.uniform(0,100)
    c=random.uniform(0,100)
    sol.append((a,b,c))
ranked_sol=[]   

for i in range(1000):
    for s in sol:
        ranked_sol.append((fitness(s[0],s[1],s[2]),s))
    ranked_sol.sort()
    ranked_sol.reverse()
    #selecting best ones  
    best_sol=ranked_sol[:100]
    print(f"this is gen {i} and we are getting best sol as- ")
    print(ranked_sol[0])
    elements=[]
    for b in best_sol:
        elements.append(b[1][0])
        elements.append(b[1][1])
        elements.append(b[1][2])
    new_gen=[]
    for i in range(500):
        a=random.choice(elements)*random.uniform(0.99,1.01)
        c=random.choice(elements)*random.uniform(0.99,1.01)
        d=random.choice(elements)*random.uniform(0.99,1.01)
        new_gen.append((a,c,d))
        

        sol=new_gen  
```

```python
#No of items there are - (Take input from User)
n=int(input())
bag_val=[]
for i in range(n):
    a=int(input())
    b=int(input())
    bag_val.append((a,b))
```

```python
#Every row as Weight,value
import numpy as np
bag_con=np.array(bag_val)
print(bag_con)
```
